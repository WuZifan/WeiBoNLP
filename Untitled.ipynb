```python
import pandas as pd
import numpy as np
import jieba as jb
import jieba.posseg as pseg
import jieba.analyse
from gensim import corpora, models
import matplotlib.pyplot as plt
```

```python
raw_data=pd.read_csv('left_11335.csv')
raw_data.head()
raw_data.shape
```

```python
raw_data.drop(['Unnamed: 0','create_time','locate','tools'],axis=1,inplace=True)
raw_data.drop(['like_nor','repost_nor','comment_nor'],axis=1,inplace=True)
```

```python
raw_data.head()
```

```python
stop_list=[]
fff=open('stop_words.txt','r')
for li in fff:
    stop_list.append(li[:-1])
fff.close()
```

# 1、提取关键词

Step1：分词&去除停用词和标点符号

    1. 思路1：通过自己进行分词-》去除停用词和标点符号-》计算TFIDF来得到关键词。
        问题：文本集不够大，计算TFIDF可能不准。
    2. 思路2：利用jieba提供接口计算TFIDF，可以通过词性来对分词筛选做控制。（前置的数据清理仅仅使用去除空格）

```python
key_words=[]
for i in range(raw_data.shape[0]):
    weibo=raw_data.loc[i,'title']
    weibo=''.join(weibo.split())
    
    k_w=jieba.analyse.extract_tags(weibo, topK=10, withWeight=False,allowPOS=('n','ns','nr',
                                                                                'nt','nz','v','a'))
    k_w=[wd for wd in k_w if str(wd) not in stop_list]
    key_words.append(k_w)
    
```

```python
# sentence='BBC推送：7点25分左右泰国曼谷市中心发生大爆炸....在市中心旅游景点四面佛附近....BBC记者表示现场伤亡严重....'
# result=jieba.analyse.extract_tags(sentence, topK=20, withWeight=False,allowPOS=('n','ns','nr',
#                                                                                 'nt','nz','v','a'))
# result
```

```python
raw_data['key_words']=key_words
```

```python
raw_data.head()
```

```python
word_dict=corpora.Dictionary(key_words)
print(len(word_dict))
corpus_list = [word_dict.doc2bow(text) for text in key_words]
print(len(corpus_list))
```

```python
lda = models.ldamodel.LdaModel(corpus=corpus_list,id2word=word_dict,num_topics=5,alpha='auto')
```

```python
for i in range(5):
    print(i,lda.print_topic(i,20))
    print()
```

```python
type_list=[]
for i in range(raw_data.shape[0]):
    k_w=key_words[i]
    temp_re=word_dict.doc2bow(k_w)
    result=lda[temp_re]
    result.sort(key=lambda x:-x[1])
    print(result)
    type_list.append(result[0][0])
```

```python
raw_data['type']=type_list
raw_data.head()
```

```python
type_list
```

```python
weibo_type_dict={}
for name,group in raw_data.groupby('type'):
    weibo_type_dict[name]=group
```

```python
comment_list=[]
like_list=[]
repost_list=[]
for key in weibo_type_dict.keys():
    temp_df=weibo_type_dict[key]
    
    temp_avg_comment=np.sum(temp_df['comment'])/temp_df.shape[0]
    temp_avg_like=np.sum(temp_df['like'])/temp_df.shape[0]
    temp_avg_repost=np.sum(temp_df['repost'])/temp_df.shape[0]
    
    comment_list.append(temp_avg_comment)
    like_list.append(temp_avg_like)
    repost_list.append(temp_avg_repost)
    
    print(key,temp_avg_comment)
#     print(key,temp_avg_like)
#     print(key,temp_avg_repost)
```

```python
plt.plot(comment_list,'r')
# plt.plot(like_list)
plt.plot(repost_list,'b')
```
